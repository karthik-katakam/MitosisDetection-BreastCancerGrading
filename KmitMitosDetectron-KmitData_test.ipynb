```python
def overlaps(boxA,boxB,iou):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
    calculated_iou = interArea / float(boxAArea + boxBArea - interArea)
    return calculated_iou>=iou
```

```python
def cal_fp_fn_tp(actual_bbox,pred_bbox,iou):
    flag=[False]*len(pred_bbox)
    fp,tp,fn=[],[],[]
    for i in range(len(actual_bbox)):
        f=0
        for j in range(len(pred_bbox)):
            if overlaps(actual_bbox[i],pred_bbox[j],iou):
                flag[j]=flag[j]|True
                f=1
        if f==0:
            fn.append(actual_bbox[i])
    for i in range(len(pred_bbox)):
        if flag[i]:
            tp.append(pred_bbox[i])
        else:
            fp.append(pred_bbox[i])
    return fp,fn,tp

```
